## Popularity Based Recommendation System

```python
# Joining books and ratings into one table
books_with_ratings = ratings.merge(books,on = 'ISBN')
```

```python
books_with_ratings.head()
```

```python
books_with_ratings.shape
```

```python
ratings.shape
```

```python
books.shape
```

```python
popular_df = books_with_ratings.groupby('Book-Title').agg(num_rating=('Book-Rating','count'),avg_rating=('Book-Rating','mean'))
```

```python
popular_df = popular_df.reset_index()
```

```python
popular_df.head()
```

```python
popular_df.sort_values('num_rating',ascending=False)
```

```python
# Popularity is based on no of people read the book ('num_rating'>300)
# It is based on rating it has got.
popular_df = popular_df[popular_df['num_rating']>300].sort_values('avg_rating',ascending=False)
```

```python
popular_df.head(50)
```

```python
# For the model deployment I need Book-title,Author,Image Url
popular_df = popular_df.merge(books,on='Book-Title').drop_duplicates('Book-Title')[['Book-Title',
                                                                                   'Book-Author',
                                                                                   'Image-URL-M',
                                                                                   'num_rating',
                                                                                   'avg_rating']]
```

# Colaborative Filtering
- Similar book prediction based on users feedback


```python
books_with_ratings.head()
```

```python
users.shape
```

```python
# Grouping based on user-id tells the no of books rated by each user
x = books_with_ratings.groupby('User-ID').count()
```

```python
x.sort_values('Book-Rating',ascending=False)
```

```python
x.shape
```

```python
# Select only users who atleast give feedback for 200 books
x = x['Book-Rating']>200
x
```

```python
power_users = x[x].index
```

```python
power_users
```

```python
 power_users.shape
```

```python
power_users.sort_values(ascending=False)
```

```python
# Selecting the records of power users
filtered_data = books_with_ratings[books_with_ratings['User-ID'].isin(power_users)]
```

```python
filtered_data
```

```python
# I am considering the best users (atleast 200 books feedback) gropu them based on book_title
y=filtered_data.groupby('Book-Title').count()
```

```python
y
```

```python
# The above dataframe tells how many users have read that book.
y.sort_values('User-ID',ascending=False)
```

```python
y = y['User-ID'] >= 50
```

```python
y
```

```python
y.shape
```

```python
famous_books = y[y].index
```

```python
famous_books
```

```python
final_ratings = filtered_data[filtered_data['Book-Title'].isin(famous_books)]
```

```python
final_ratings
```

```python
# Pivot Table giving the ratings for each book from each user
# Books rows with userid as column
pt = final_ratings.pivot_table(index='Book-Title',columns='User-ID',values='Book-Rating')
```

```python
pt
```

```python
pt = pt.fillna(0)
```

```python
pt
```

```python
similarity_scores = cosine_similarity(pt)
```

```python
similarity_scores
```

```python
df_temp = pd.DataFrame(similarity_scores)
```

```python
df_temp
```

```python
df_temp_name=df_temp
```

```python
df_temp_name.index=pt.index
df_temp_name.columns = pt.index
```

```python
df_temp_name
```

```python
def recommend(book_name):
    index = np.where(pt.index == book_name)[0][0]
    similar_items = sorted(list(enumerate(similarity_scores[index])),key=lambda x: x[1],reverse=True)[1:6]
    #lets create empty list and in that lies it should contain book-author book-title image-url
    data=[]
    for i in similar_items:
        item=[]
        temp_df=books[books['Book-Title'] == pt.index[i[0]]]
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Title'].values))
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Author'].values))
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Image-URL-M'].values))
        data.append(item)
    return data
```

```python
recommend('1984')
```

```python
recommend('The Hours : A Novel')
```
